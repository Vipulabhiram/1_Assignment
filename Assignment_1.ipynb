```python
#first we are installing transformer from it we are importing pipeline function
!pip install transformers
from transformers import pipeline
```

```python
question_and_answerer = pipeline("question-answering",model="HuggingFaceH4/zephyr-7b-beta")
```

```python
context = """India is a country located in South Asia. It is the seventh-largest country by land area and the most populous country in the world. India has a rich history and diverse culture, with hundreds of languages, religions, and ethnic groups. The country gained independence from British rule in 1947 and became a republic in 1950. The capital city of India is New Delhi, while its largest
 city is Mumbai. India has a federal parliamentary system and is known for its growing economy, space research, and contributions to science, technology, and the arts.
"""
```

```python
question = "Tell me about India?"
result = question_and_answerer(
    question=question,
    context=context
)["answer"]
```

```python
print(result)
```
